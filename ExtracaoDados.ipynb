```python
import pandas as pd
```

```python
df = pd.read_csv("data/ESCOLAS_POA_2018_.csv", encoding='cp1252', sep="|")
```

```python
df_escola = df.loc[:, ['CO_ENTIDADE', 'NO_ENTIDADE', 'TP_DEPENDENCIA', 'TP_ATIVIDADE_COMPLEMENTAR', 'IN_ALIMENTACAO', 'IN_COMUM_CRECHE', 'IN_COMUM_PRE', 'IN_COMUM_FUND_AI', 'IN_COMUM_FUND_AF', 'IN_COMUM_MEDIO_MEDIO']]
```

```python
df_escola2 = df_escola
df_escola2.loc[df_escola2.TP_DEPENDENCIA == 1, 'TIPO'] = 'Federal' 
df_escola2.loc[df_escola2.TP_DEPENDENCIA == 2, 'TIPO'] = 'Estadual'
df_escola2.loc[df_escola2.TP_DEPENDENCIA == 3, 'TIPO'] = 'Municipal'
df_escola2.loc[df_escola2.TP_DEPENDENCIA == 4, 'TIPO'] = 'Privada'
df_escola2.loc[df_escola2.TP_ATIVIDADE_COMPLEMENTAR == 1, 'ATIVIDADE_COMPLEMENTAR'] = 'Sim'
df_escola2.loc[df_escola2.TP_ATIVIDADE_COMPLEMENTAR == 0, 'ATIVIDADE_COMPLEMENTAR'] = 'Não'
df_escola2.loc[df_escola2.IN_ALIMENTACAO == 1, 'ALIMENTACAO'] = 'Sim'
df_escola2.loc[df_escola2.IN_ALIMENTACAO == 0, 'ALIMENTACAO'] = 'Não'
df_escola2.loc[df_escola2.IN_COMUM_CRECHE == 1, 'CRECHE'] = 'Sim'
df_escola2.loc[df_escola2.IN_COMUM_CRECHE == 0, 'CRECHE'] = 'Não'
df_escola2.loc[df_escola2.IN_COMUM_PRE == 1, 'PRE_ESCOLA'] = 'Sim'
df_escola2.loc[df_escola2.IN_COMUM_PRE == 0, 'PRE_ESCOLA'] = 'Não'
df_escola2.loc[df_escola2.IN_COMUM_FUND_AI == 1, 'FUNDAMENTAL_ANOS_INICIAIS'] = 'Sim'
df_escola2.loc[df_escola2.IN_COMUM_FUND_AI == 0, 'FUNDAMENTAL_ANOS_INICIAIS'] = 'Não'
df_escola2.loc[df_escola2.IN_COMUM_FUND_AF == 1, 'FUNDAMENTAL_ANOS_FINAIS'] = 'Sim'
df_escola2.loc[df_escola2.IN_COMUM_FUND_AF == 0, 'FUNDAMENTAL_ANOS_FINAIS'] = 'Não'
df_escola2.loc[df_escola2.IN_COMUM_MEDIO_MEDIO == 1, 'MEDIO'] = 'Sim'
df_escola2.loc[df_escola2.IN_COMUM_MEDIO_MEDIO == 0, 'MEDIO'] = 'Não'
```

```python
df_escola3 = df_escola2.drop(['TP_DEPENDENCIA', 'TP_ATIVIDADE_COMPLEMENTAR', 'IN_ALIMENTACAO', 'IN_COMUM_CRECHE', 'IN_COMUM_PRE', 'IN_COMUM_FUND_AI', 'IN_COMUM_FUND_AF', 'IN_COMUM_MEDIO_MEDIO'], axis=1)
```

```python
df_avaliacao2 = pd.merge(df_escola3, df_avaliacao, how = 'inner', on = ['NO_ENTIDADE'])
```

```python
df_avaliacao3 = df_avaliacao2.drop(['TIPO', 'ATIVIDADE_COMPLEMENTAR', 'ALIMENTACAO', 'CRECHE', 'PRE_ESCOLA', 'FUNDAMENTAL_ANOS_INICIAIS', 'FUNDAMENTAL_ANOS_FINAIS', 'MEDIO'], axis=1)
```

```python
import sqlite3
from pandas import DataFrame
```

```python
conn = sqlite3.connect('TestDB2.sqlite')
c = conn.cursor()

c.execute('CREATE TABLE AVALIACAO (NO_ENTIDADE text, IDEB_ANOS_INICIAIS number, IDEB_ANOS_FINAIS number)')
c.execute('CREATE TABLE CONTATO (CO_ENTIDADE number, NU_DDD number, NU_TELEFONE_PUBLICO number, email text, url_website)')
c.execute('CREATE TABLE LOCALIZACAO (CO_ENTIDADE number, latitude number, longitude number, DS_ENDERECO text, NU_ENDERECO number, DS_COMPLEMENTO text, NO_BAIRRO text, CO_CEP number)')
c.execute('CREATE TABLE ACESSIBILIDADE (CO_ENTIDADE number, IN_SALA_ATENDIMENTO_ESPECIAL number, IN_BANHEIRO_PNE number, IN_DEPENDENCIAS_PNE number, TP_AEE number, IN_ESPECIAL_EXCLUSIVA number)')
c.execute('CREATE TABLE INFRA (CO_ENTIDADE number, IN_LABORATORIO_INFORMATICA number, IN_QUADRA_ESPORTES_COBERTA number, IN_QUADRA_ESPORTES_DESCOBERTA number, IN_BIBLIOTECA number, IN_SALA_LEITURA number, IN_PARQUE_INFANTIL number, IN_REFEITORIO number, IN_PATIO_COBERTO number, IN_PATIO_DESCOBERTO number, IN_AREA_VERDE number, IN_INTERNET number)')
c.execute('CREATE TABLE ESCOLA (CO_ENTIDADE number, NO_ENTIDADE text, TIPO text, CRECHE text, PRE_ESCOLA text, FUNDAMENTAL_ANOS_INICIAIS text, FUNDAMENTAL_ANOS_FINAIS text, MEDIO text, ATIVIDADE_COMPLEMENTAR text, ALIMENTACAO text)')
conn.commit()

df_avaliacao3.to_sql('AVALIACAO', conn, if_exists='replace', index = False)
df_contato.to_sql('CONTATO', conn, if_exists='replace', index = False)
df_localizacao.to_sql('LOCALIZACAO', conn, if_exists='replace', index = False)
df_acessibilidade.to_sql('ACESSIBILIDADE', conn, if_exists='replace', index = False)
df_infra.to_sql('INFRA', conn, if_exists='replace', index = False)
df_escola3.to_sql('ESCOLA', conn, if_exists='replace', index = False)
```
